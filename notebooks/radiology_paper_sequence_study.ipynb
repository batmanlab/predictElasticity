```python
sequences = ['t1_pre_water', 't1_pre_in',  't1_pre_out',  't1_pre_fat', 't2', 
              't1_pos_0_water',  't1_pos_70_water', 't1_pos_160_water', 't1_pos_300_water', 'best_4',
             'all_9', 'best_4_no_clin', 'best_4_class_only', 'clin_only_class_only']
thresholds = ['2.88', '3.54', '3.77', '4.09']

multi_index = pd.MultiIndex.from_product([sequences, thresholds], names=['sequence', 'threshold'])
df_thresh = pd.DataFrame(index=multi_index, columns=['sens', 'sens_std', 'spec', 'spec_std', 'acc', 'acc_std', 'auroc', 'auroc_std'])
df_r2 = pd.DataFrame(index=sequences, columns=['r2_subj', 'r2_subj_std', 'r2_pixel', 'r2_pixel_std'])
df_dice = pd.DataFrame(index=multi_index, columns=['dice'])
```

```python
df_thresh
```

```python
# Roc Curves
fig, ax = plt.subplots(2,2, figsize=(15,15), sharex=True, sharey=True)
ax = np.ravel(ax)
for seq in sequences:
#for seq in ['best_4']:
    print(seq)
    # Dice
    ds_dice = ModelCompareDice(ds_models, do_cor=True, pred=seq).ds
    true_vals = ds_dice['image_mre'].sel(mre_type='true').values.flatten()
    true_vals = true_vals[~np.isnan(true_vals)]
    pred_vals = ds_dice['image_mre'].sel(mre_type='pred').values.flatten()
    pred_vals = pred_vals[~np.isnan(pred_vals)]
    for i,thresh in enumerate(thresholds):
        roc_output = roc_curves(df, pred=seq, label=seq, threshold=float(thresh), ax=ax[i])
        df_thresh.loc[(seq, thresh)] = roc_output
        
        # Dice
        true_thresh = np.where(true_vals>float(thresh)*1000, 1, 0)
        pred_thresh = np.where(pred_vals>float(thresh)*1000, 1, 0)
        dice = np.sum(pred_thresh[true_thresh==1])*2.0 / (np.sum(pred_thresh) + np.sum(true_thresh))
        print(f'dice ({thresh}): {dice}')
        
        df_dice.loc[(seq, thresh)] = dice

    r2_output = radiology_cor_plots(ds_models, pred=seq, do_cor=True, plot=False)
    df_r2.loc[seq] = r2_output
handles, labels = ax[0].get_legend_handles_labels()
ax[1].legend(handles, labels, loc='lower right', fontsize=14)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```
